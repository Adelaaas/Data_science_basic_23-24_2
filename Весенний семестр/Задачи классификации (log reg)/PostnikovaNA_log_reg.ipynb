Большое количество пропусков присутствует только в столбце Arrival Delay in Minutes. Так как пропусков здесь меньше 50% от всех значений в столбце, то этот столбец нет надобности удалять. Удобнее всего заменить пропуски средним значением по всему столбцу. В остальных столбцах по одному пропуску, значит можно удалить строчки с пропусками

```python
value = df['Arrival Delay in Minutes'].mean()
df['Arrival Delay in Minutes'] = df['Arrival Delay in Minutes'].fillna(value)
```

```python
df = df.dropna(axis = 0, how ='any')
```

в датайрейме есть категориальные переменные: Gender,	Customer Type, Type of Travel,	Class, satisfaction. Так как переменная satisfaction это наш У, то мы должны ее и остальные переменные закодировать.  

```python
print(df['Gender'].value_counts())
print(df['Customer Type'].value_counts())
print(df['Type of Travel'].value_counts())
print(df['Class'].value_counts())
print(df['satisfaction'].value_counts())
```

Во всех категориальных переменных макисмум 3 различных значения, значит мы можем закодировать их  с момощью Label_encoder

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
label_encoder_1 = preprocessing.LabelEncoder()
label_encoder_2 = preprocessing.LabelEncoder()
label_encoder_3 = preprocessing.LabelEncoder()
label_encoder_4 = preprocessing.LabelEncoder()
label_encoder.fit(df['Gender'])
label_encoder_1.fit(df['Customer Type'])
label_encoder_2.fit(df['Type of Travel'])
label_encoder_3.fit(df['Class'])
label_encoder_4.fit(df['satisfaction'])


df['Gender'] = label_encoder.transform(df['Gender'])
df['Customer Type'] = label_encoder_1.transform(df['Customer Type'])
df['Type of Travel'] = label_encoder_2.transform(df['Type of Travel'])
df['Class'] = label_encoder_3.transform(df['Class'])
df['satisfaction'] = label_encoder_4.transform(df['satisfaction'])
df.head()
```
