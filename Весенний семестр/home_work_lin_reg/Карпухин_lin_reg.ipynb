```python
X = df.drop(columns=['Life expectancy ', 'Country', 'Status', 'Year', 'Measles ', 'Population', 'GDP'])

Y = df['Life expectancy ']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=1/4, random_state=34)

print(f'Train dataset size: {X_train.shape}, {Y_train.shape}')
print(f'Test dataset size: {X_test.shape}, {Y_test.shape}')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

model = LinearRegression()

model.fit(X_train, Y_train)
```

```python
print('Веса всех признаков (w1, ..., w13): ', model.coef_)
print('Свободный коэффицент уравнения w0:', model.intercept_)
```

Уравнение:  53.345060906311225 + -0.451430 * HIV/AIDS + -0.073320	* thinness 5-9 years + -0.065969	* under-five deaths + -0.061283	* Alcohol + -0.016273	* Adult Mortality + -0.004785	* Hepatitis B + 0.000453	* percentage expenditure + 0.005999	* Polio + 0.008860	* thinness 1-19 years + 0.016934 *	Diphtheria + 0.030864	* BMI + 0.074996	* Total expenditure + 0.086997	* infant deaths + 0.858690	* Schooling + 10.087647	* Income composition of resources

```python
features = X_train.columns

coeff_df = pd.DataFrame(model.coef_, columns=['Coefficient'])
coeff_df['features'] = features

coeff_df.sort_values(by='Coefficient')
```

признак GDP имеет сильно маленький коэфицент, удаляем

самый значимый признак - Income composition of resources, значит больше всего на ожидаемую продолжительность жизни (Life expectancy) влияет доходная структура ресурсов (Income composition of resources)

```python
y_pred = model.predict(X_test)

print('Mean Absolute Error:', mean_absolute_error(Y_test, y_pred))
print('Mean Squared Error:', mean_squared_error(Y_test, y_pred))
print('R2 score:', r2_score(Y_test, y_pred))
```
