```python
rr = pd.read_csv('/content/input (2).csv')
vr = rr[rr['ДИСЦИПЛИНА'] == 'Компьютерные технологии в дизайне']
vr.drop(['НАПРАВЛЕНИЕ'], axis = 1,inplace=True)
vr.drop(['STD_ID'], axis = 1,inplace=True)
vr.drop(['КУРС'], axis = 1,inplace=True)
vr.drop(['СЕМЕСТР'], axis = 1,inplace=True)
vr.drop(['АТТЕСТАЦИЯ'], axis = 1,inplace=True)
vr
```

```python
s5 = vr.sort_values(by = ['ГОД'],ascending = True,ignore_index = True)#Сортировка таблицы по заданному столбцу, ascending = True/False отвечает за сортировку по возрастанию/убыванию
s5
```

```python
s6 = s5.groupby(['ГОД']).agg({'mark': 'mean'})
s6
```
