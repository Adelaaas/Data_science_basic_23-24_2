```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('train_var_3.csv')
df.head()
```

# Новый раздел

Задание 1


```python
y = df[df['MaxTemp'].isnull()==False]
y
```

```python
x = df['MinTemp']
y = df['MaxTemp']

fig, ax = plt.subplots(1, 2)

ax[0].scatter(x, y, c = 'deeppink')
ax[0].set_xlabel('MinTemp')
ax[0].set_ylabel('MaxTemp')

ax[1].scatter(y, x, c = 'orange')
ax[1].set_xlabel('MaxTemp')
ax[1].set_ylabel('MinTemp')
fig.suptitle('Dependence MinTemp from MaxTemp and conversaly')

plt.show()
```

Точки на обоих диаграммах расположены по одной стороне от линии, которую можно провести, и они находятся недалеко друг от друга и между ними прямая пропорционалная связь

Задание 3


```python
df = df.dropna()
dfend= df.select_dtypes(exclude = ['object'])
dfend.head()
```

```python
fig, axes = plt.subplots(2, 9, figsize=(20,10))
sns.boxplot(dfend['MinTemp'], ax=axes[0,0], width=0.3).set_title('MinTemp')
sns.boxplot(dfend['MaxTemp'], ax=axes[0,1], width=0.3).set_title('MaxTemp')
sns.boxplot(dfend['Rainfall'], ax=axes[0,2], width=0.3).set_title('Rainfall')
sns.boxplot(dfend['Evaporation'], ax=axes[0,3], width=0.3).set_title('Evaporation')
sns.boxplot(dfend['Sunshine'], ax=axes[0,4], width=0.3).set_title('Sunshine')
sns.boxplot(dfend['WindGustSpeed'], ax=axes[0,5], width=0.3).set_title('WindGustSpeed')
sns.boxplot(dfend['WindSpeed9am'], ax=axes[0,6], width=0.3).set_title('WindSpeed9am')
sns.boxplot(dfend['WindSpeed3pm'], ax=axes[0,7], width=0.3).set_title('WindSpeed3pm')
sns.boxplot(dfend['Humidity9am'], ax=axes[0,8], width=0.3).set_title('Humidity9am')
sns.boxplot(dfend['Humidity3pm'], ax=axes[1,0], width=0.3).set_title('Humidity3am')
sns.boxplot(dfend['Pressure9am'], ax=axes[1,1], width=0.3).set_title('Pressure9am')
sns.boxplot(dfend['Pressure3pm'], ax=axes[1,2], width=0.3).set_title('Pressure9am')
sns.boxplot(dfend['Cloud9am'], ax=axes[1,3], width=0.3).set_title('Cloud9am')
sns.boxplot(dfend['Cloud3pm'], ax=axes[1,4], width=0.3).set_title('Cloud3pm')
sns.boxplot(dfend['Temp9am'], ax=axes[1,5], width=0.3).set_title('Temp9am')
sns.boxplot(dfend['Temp3pm'], ax=axes[1,6], width=0.3).set_title('Temp3pm')
sns.boxplot(dfend['RainToday'], ax=axes[1,7], width=0.3).set_title('RainToday')
sns.boxplot(dfend['RainTomorrow'], ax=axes[1,8], width=0.3).set_title('RainTomorrow')
```

Количество различных значений в столбе Rainfall наиболшее количество.

```python
plt.figure(figsize=(13,13), dpi= 80)
sns.heatmap(df.corr(), xticklabels=df.corr().columns, yticklabels=df.corr().columns, cmap='RdYlGn', center=0, annot=True)
plt.title('Correlogram', fontsize=22)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.show()
```

Наиболее высокая коррелляция у столбцов MaxTemp и Temp3pm;
наиболее низкая коррелляция у столбцов MinTemp и Temp9am
