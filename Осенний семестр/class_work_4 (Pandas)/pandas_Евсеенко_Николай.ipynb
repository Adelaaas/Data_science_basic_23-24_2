```python
# создадим тестовый датасет
df = pd.DataFrame({
    'имя': ['Петя', 'Вася', 'Аня', 'Света'],
    'возраст': [45, 26, 41 , 21],
    'доход': [100000, 80000, 150000, 60000],
    'надежность клиента (0..1)': [0.65, 0.74, 0.87, 0.68],
    'пол': ['муж', 'муж', 'жен', 'жен']},
    index=[0, 1, 2, 3])

df

# ваш код тут
df[df['возраст'] >= 40]
```

```python
s=df['доход'].mean()
df[df['доход']>s]
```

```python
k=df['надежность клиента (0..1)'].mean()
df[(df['доход']>s)&(df['надежность клиента (0..1)']<k)]
```

```python
df['важность клиента'] = df['надежность клиента (0..1)']*df['доход']
df
```

```python
s=df['возраст'].mean()
df['возможная долгосрочность клиента']=df['важность клиента']*s
df
```
