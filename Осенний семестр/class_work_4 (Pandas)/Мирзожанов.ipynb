```python
#Task 2
df = pd.DataFrame({
    'имя': ['Петя', 'Вася', 'Аня', 'Света'],
    'возраст': [45, 26, 41 , 21],
    'доход': [100000, 80000, 150000, 60000],
    'надежность клиента (0..1)': [0.65, 0.74, 0.87, 0.68],
    'пол': ['муж', 'муж', 'жен', 'жен']},
    index=[0, 1, 2, 3])
df
```

```python
df[df['возраст'] > 40]
```

```python
sr_dohod = df["доход"].mean()
#print(sr_dohod)
df[df["доход"] > sr_dohod]
```

```python
sr_dohod = df["доход"].mean()
sr_nad = df["надежность клиента (0..1)"].mean()
#print(sr_nad)
df[(df["доход"] > sr_dohod) & (df["надежность клиента (0..1)"] < sr_nad)]
```

```python
values = df["надежность клиента (0..1)"] * df["доход"]
df['важность клиента'] = values
df
```

```python
sr_old = df['возраст'].mean()
qwerty = df['важность клиента'] * (sr_old - df["возраст"])
df['возможная долгосрочность клиента'] = qwerty
df
```
