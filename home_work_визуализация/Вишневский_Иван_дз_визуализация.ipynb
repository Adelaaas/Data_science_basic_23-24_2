```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("train_var_1.csv")
df.head()
```

```python
fig, ax = plt.subplots()

ax.scatter(df['Vmag'], df['e_Plx'], c = 'brown')

plt.show()
```

```python
plt.figure(figsize=(15,7))

c = sns.cubehelix_palette(as_cmap=True, light=.7)

sns.heatmap(df.isna().transpose(),
            cbar_kws={'label': 'Missing Data'})
```

```python
df[['Vmag', 'Plx', 'e_Plx', 'B-V', 'Amag', 'TargetClass']].plot(kind='box')
```

```python
fig, axes = plt.subplots(1, 6, figsize=(15,6))
sns.boxplot(df['Vmag'], ax=axes[0], width=0.3)
sns.boxplot(df['Plx'], ax=axes[1], width=0.3)
sns.boxplot(df['e_Plx'], ax=axes[2], width=0.3)
sns.boxplot(df['B-V'], ax=axes[3], width=0.3)
sns.boxplot(df['Amag'], ax=axes[4], width=0.3)
sns.boxplot(df['TargetClass'], ax=axes[5], width=0.3)
```

```python
sns.barplot(x='TargetClass', y=df['TargetClass'])
```

```python
plt.figure(figsize=(12,10), dpi= 80)
sns.heatmap(df.corr(), xticklabels=df.corr().columns, yticklabels=df.corr().columns, cmap='RdYlGn', center=0, annot=True)

plt.title('Correlogram of mtcars', fontsize=22)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10,8), dpi= 80)
sns.pairplot(df, kind="reg", hue="species")
plt.show()
```
