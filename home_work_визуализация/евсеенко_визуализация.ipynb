```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('train_var_2.csv')
df.head()
```

```python
fig, ax = plt.subplots()
x=df['age']
y=df['balance']
ax.scatter(x,y, c='deeppink')
plt.xlabel("age")
plt.ylabel("balance")
```

До 20 лет денег у людей практически нет. От 30 до 60 у людей примерно по 20 тысяч. После 60 лет денег становится меньше, скорее всего из-за выхода на пенсию. В то же время, как богатые люди не зависят от возраста.


```python
import seaborn as sns
g=df.replace('unknown', None)
g.isna().sum()

plt.figure(figsize=(15,7))

cmap = sns.cubehelix_palette(as_cmap=" ", light=.9)

sns.heatmap(g.isna().transpose(), cmap=cmap,
            cbar_kws={'label': 'Missing Data'})

```

По графику видно, что больше всего пропущенных значений в poutcome и contact.

```python


fig, axes = plt.subplots(2, 4, figsize=(17,10))
sns.boxplot(df['age'], ax=axes[0,0], width=0.3).set_title('age')
sns.boxplot(df['balance'], ax=axes[0,1], width=0.3).set_title('balance')
sns.boxplot(df['day'], ax=axes[0,2], width=0.3).set_title('day')
sns.boxplot(df['duration'], ax=axes[1,0], width=0.3).set_title('duration')
sns.boxplot(df['campaign'], ax=axes[1,1], width=0.3).set_title('campaign')
sns.boxplot(df['pdays'], ax=axes[1,2], width=0.3).set_title('pdays')
sns.boxplot(df['previous'], ax=axes[1,3], width=0.3).set_title('previous')

```

Вывод:  очень много выбросов в боксе по балансу, можно судить о том, что большая часть населения живёт бедно

```python

x = df['deposit'].unique()
y = df['deposit'].value_counts()
sns.barplot(x=x, y=y).set_title('deposit')
```

Вывод: в целом можно сказать, что людей без депозита больше
