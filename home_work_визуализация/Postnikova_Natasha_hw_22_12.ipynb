```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('train_var_3.csv')
df.head()
```

Постройте Диаграмму разброса (точечный график) от двух переменных: MinTemp и MaxTemp.

```python
y = df[df['MaxTemp'].isnull()==False]#df['MaxTemp']
y
```

```python
x = df['MinTemp']
y = df['MaxTemp']

fig, ax = plt.subplots(1, 2)

ax[0].scatter(x, y, c = 'deeppink')
ax[0].set_xlabel('MinTemp')
ax[0].set_ylabel('MaxTemp')

ax[1].scatter(y, x, c = 'orange')
ax[1].set_xlabel('MaxTemp')
ax[1].set_ylabel('MinTemp')
fig.suptitle('Dependence MinTemp from MaxTemp and conversaly')
#ax[0].set_facecolor('black')
#fig.set_facecolor('r')
plt.show()
```

Вывод: Т.к. точки на обоих графиков находятся по одну сторону от прямой, которую можно провести, и они рсаположены близко друг к другу, можно сказать, что эти две температуры взаимосвязаны между собой, и взаимосвязаны положительно. Таким образом, когда одна температура растет, растет и другая.

Построить boxplot для всех числовых переменных. График должен быть один, используя разные subplot() для каждой переменной (т.е. одна фигура и несколько сабплотов).

```python
#df.select_dtypes(include = ['int64'])
df = df.dropna()
df1= df.select_dtypes(exclude = ['object'])
df1.head()
```

```python
fig, axes = plt.subplots(2, 9, figsize=(20,10))
sns.boxplot(df1['MinTemp'], ax=axes[0,0], width=0.3).set_title('MinTemp')
sns.boxplot(df1['MaxTemp'], ax=axes[0,1], width=0.3).set_title('MaxTemp')
sns.boxplot(df1['Rainfall'], ax=axes[0,2], width=0.3).set_title('Rainfall')
sns.boxplot(df1['Evaporation'], ax=axes[0,3], width=0.3).set_title('Evaporation')
sns.boxplot(df1['Sunshine'], ax=axes[0,4], width=0.3).set_title('Sunshine')
sns.boxplot(df1['WindGustSpeed'], ax=axes[0,5], width=0.3).set_title('WindGustSpeed')
sns.boxplot(df1['WindSpeed9am'], ax=axes[0,6], width=0.3).set_title('WindSpeed9am')
sns.boxplot(df1['WindSpeed3pm'], ax=axes[0,7], width=0.3).set_title('WindSpeed3pm')
sns.boxplot(df1['Humidity9am'], ax=axes[0,8], width=0.3).set_title('Humidity9am')
sns.boxplot(df1['Humidity3pm'], ax=axes[1,0], width=0.3).set_title('Humidity3am')
sns.boxplot(df1['Pressure9am'], ax=axes[1,1], width=0.3).set_title('Pressure9am')
sns.boxplot(df1['Pressure3pm'], ax=axes[1,2], width=0.3).set_title('Pressure9am')
sns.boxplot(df1['Cloud9am'], ax=axes[1,3], width=0.3).set_title('Cloud9am')
sns.boxplot(df1['Cloud3pm'], ax=axes[1,4], width=0.3).set_title('Cloud3pm')
sns.boxplot(df1['Temp9am'], ax=axes[1,5], width=0.3).set_title('Temp9am')
sns.boxplot(df1['Temp3pm'], ax=axes[1,6], width=0.3).set_title('Temp3pm')
sns.boxplot(df1['RainToday'], ax=axes[1,7], width=0.3).set_title('RainToday')
sns.boxplot(df1['RainTomorrow'], ax=axes[1,8], width=0.3).set_title('RainTomorrow')
```

Вывод: По графику видно, что больше всего выбросов у нас находится в столбце RainFall, т.е. количество различных значений там больше всего. Также видно, что больше помежуток разброса значений в столбцах Humidity9am, Humidity3pm

Построить матрицу корреляции для всех переменных.

```python
df.corr()
```

```python
plt.figure(figsize=(13,13), dpi= 80)
sns.heatmap(df.corr(), xticklabels=df.corr().columns, yticklabels=df.corr().columns, cmap='RdYlGn', center=0, annot=True)
plt.title('Correlogram of mtcars', fontsize=22)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.show()
```

Вывод: По кореллограмме видно, что наиболее высокая коррелляция у столбцов MaxTemp и Temp3pm, т.е. днем температура обычно самая высокая, а также у столбцов MinTemp и Temp9am, т.е. утром температура самая низкая. Еще высокая коррелляция у столбцов Pressure3pm и Pressure9am, значит давление зависит от давления которое было в другое время проведения измерения.
  Наиболее отрицательная корреляция у столбцов Cloud3pm и Sunshine, т.е. днем когда на небе солнце почти не бывает облачно и наоборот. также корреляция наиболее отрицательна у столбцов Cloud9am и Sunshine, т.е. с утра тоже маловерятно может быть солнце и облака одновременно, но все же более вероятно, чем днем. А одна из нейтральных корреляций - это взаимосвязь столбцов Sunshine и WindSpeed3pm.

Построить графики парных зависимостей между всеми числовыми переменными.

```python
plt.figure(figsize=(20,15), dpi= 80)
sns.pairplot(df1, kind="reg")
plt.show()
```

Вывод: В целом все параметры связаны друг с другом. Это можно судить по точеным графикам зависимости одного параметра от другого. Все значения располгаются примерно в одном диапазоне, какие-то аномальные значения в большинстве своем отсутсвтуют.
