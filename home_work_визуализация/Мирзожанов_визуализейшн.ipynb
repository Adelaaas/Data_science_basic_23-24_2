```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Task 1
```

```python
df = pd.read_csv("/content/train_var_2.csv")
df.head()
```

```python
x = df['age']
y = df['balance']
fig, ax = plt.subplots()
ax.scatter(x, y, c = 'deeppink')
plt.title("Зависимость баланса от возраста")
plt.xlabel("Возраст")
plt.ylabel("Баланс")
plt.grid()
plt.show()

```

**Вывод:** у людей младше 20 лет на счету почти нет денег, что и ожидаемо.У людей среднего возраста имеется некая суммая на счетах, это показывает резкий подъем значения баланса у людей в возрасте от 25 до 60.

```python
fig, axes = plt.subplots(2, 4, figsize=(17,10))
sns.boxplot(df['age'], ax=axes[0,0], width=0.3).set_title('age')
sns.boxplot(df['balance'], ax=axes[0,1], width=0.3).set_title('balance')
sns.boxplot(df['day'], ax=axes[0,2], width=0.3).set_title('day')
sns.boxplot(df['duration'], ax=axes[1,0], width=0.3).set_title('duration')
sns.boxplot(df['campaign'], ax=axes[1,1], width=0.3).set_title('campaign')
sns.boxplot(df['pdays'], ax=axes[1,2], width=0.3).set_title('pdays')
sns.boxplot(df['previous'], ax=axes[1,3], width=0.3).set_title('previous')

```

**Вывод:** Достаточно много выбросов в боксе по возрасту, можно сказать, что много старого населения. Также очень много выбросов в боксе по балансу, можно мудить о том, что большая часть населения живёт бедно

```python
df.corr()
```

```python
plt.figure(figsize=(13,13), dpi= 80)
sns.heatmap(df.corr(), xticklabels=df.corr().columns, yticklabels=df.corr().columns, cmap='RdYlGn', center=0, annot=True)
plt.title('Correlogram of mtcars', fontsize=22)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.show()
```

**Вывод:** из матрицы видно, что ни один из типов данных практически не коррелирует с другим. Это означает, что данные не зависят друг от друга, что очевидно также по названиям столбцов
