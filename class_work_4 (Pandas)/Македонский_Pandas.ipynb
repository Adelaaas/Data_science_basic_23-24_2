```python
[2]
df = pd.DataFrame({
    'имя': ['Петя', 'Вася', 'Аня', 'Света'],
    'возраст': [45, 26, 41 , 21],
    'доход': [100000, 80000, 150000, 60000],
    'надежность клиента (0..1)': [0.65, 0.74, 0.87, 0.68],
    'пол': ['муж', 'муж', 'жен', 'жен']},
    index=[0, 1, 2, 3])

df[df['возраст'] > 40]



```

```python
mean_income = df['доход'].mean()
df[df['доход'] > mean_income]

```

```python
mean_reliability = df['надежность клиента (0..1)'].mean()
df[(df['доход'] > mean_income) & (df['надежность клиента (0..1)'] < mean_reliability)]

```

```python
df['важность клиента'] = df['доход'] * df['надежность клиента (0..1)']
df
```

```python
meanage = df['возраст'].mean()
df['возможная долгосрочность клиента'] = df['важность клиента'] * (meanage - df['возраст'])
df
```
